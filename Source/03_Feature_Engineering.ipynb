# Feature Engineering

```python
import pandas as pd
```

### Load Data from U.S. Fish & Wildlife Service

```python
species = pd.read_pickle("../Data/fws.pkl")
species.head()
```

### Create Dummy Variables for Categorical Data

```python
#create dummy variables
species = pd.get_dummies(data=species, columns=['Group', 'Region', 'VIP', 'State'])
print(species.shape)
species.head()
```

### Data Loading

```python
species.to_pickle("../Data/species.pkl")
```
